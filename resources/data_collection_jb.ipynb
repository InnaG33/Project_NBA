```python
#dependencies
import pandas as pd
import os
import csv
```

```python
player_data = 'player_data.csv'
player_data = pd.read_csv(player_data)
player_data
player_data['name']
```

```python
player_data = player_data.rename(columns={'name': 'Player'})
player_data
```

```python
players = 'Players.csv'
players = pd.read_csv(players)
players
players['Player']
```

```python
complete_player_data = pd.merge(player_data, players,
                                 how='left', on=["Player","Player"])
complete_player_data
```

```python
# sorting by first name 
complete_player_data.sort_values("Player", inplace = True) 
  
# dropping ALL duplicte values 
complete_player_data.drop_duplicates(subset ="Player", 
                     keep = False, inplace = True) 
  
# displaying data 
complete_player_data
```

```python
clean_player_data = complete_player_data.dropna(axis = 0, how ='any') 
  
# comparing sizes of data frames 
print("Old data frame length:", len(complete_player_data), "\nNew data frame length:",  
       len(clean_player_data), "\nNumber of rows with at least 1 NA value: ", 
       (len(complete_player_data)-len(clean_player_data)))
```

```python
clean_player_data
```

```python
#list all columns and create new columns I want (height and weight)
list(clean_player_data.columns)

new_columns = [
    'Player',
    'height_x',
    'weight_x',
    'height_y',
    'weight_y',
]
```

```python
#new df with just high and weight
height_weight_df = clean_player_data[new_columns]
height_weight_df.head()
```

```python
new_columns_1 = [
    'Player',
    'height_y',
    'weight_y',
]

clean_height_weight_df = clean_player_data[new_columns_1]
clean_height_weight_df.head()
```

```python
stats_data = 'Seasons_Stats.csv'
stats_data = pd.read_csv(stats_data)
stats_data
```

```python
stats_data.loc[stats_data["Player"] == "Alaa Abdelnaby"]
```

```python
list(stats_data.columns)

new_columns = [
    'Player',
    'PTS',
]

# new df with just player and pts
PTS_df = stats_data[new_columns]
PTS_df.head()
```

```python
PTS_df.loc[PTS_df["Player"] == "Alaa Abdelnaby"]
```

```python
len(PTS_df)
```

```python
PTS_height_weight_data = pd.merge(clean_height_weight_df, PTS_df,
                                 how='left', on=["Player","Player"])
PTS_height_weight_data
```

```python
len(PTS_height_weight_data)
```

```python
PTS_height_weight_data.loc[PTS_height_weight_data["Player"] == "Alaa Abdelnaby"]
```

```python
player_height = PTS_height_weight_data.groupby("Player").mean()["height_y"]
player_height

player_weight = PTS_height_weight_data.groupby("Player").mean()["weight_y"]
player_weight
```

```python
PTS_count = PTS_height_weight_data.groupby("Player").sum()["PTS"]
PTS_count
```

```python
PTS_summary = pd.DataFrame({"Total PTS": PTS_count,
                            "Height": player_height,
                          })
PTS_summary
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(player_height,PTS_count)
plt.title("Height vs. PTS")
plt.xlabel('Height (cm)')
plt.ylabel('PTS Total')
plt.show()
```

```python
plt.scatter(player_weight,PTS_count)
plt.title("Weight vs. PTS")
plt.xlabel('Weight (kg)')
plt.ylabel('PTS Total')
plt.show()
```

```python
import scipy.stats as st
```

```python
#calculating correlation between height and PTS
correlation = round(st.pearsonr(player_height,PTS_count)[0],2)
print(f"The correlation between both factors is {correlation}")
```

```python
#calculating correlation between weight and PTS
correlation = round(st.pearsonr(player_weight,PTS_count)[0],2)
print(f"The correlation between both factors is {correlation}")
```

```python
from scipy.stats import linregress
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(player_height,PTS_count)
regress_values = player_height * slope + intercept

plt.scatter(player_height,PTS_count)
plt.plot(player_height,regress_values, color='r')
plt.title("Height vs. PTS")
plt.xlabel('Height (cm)')
plt.ylabel('PTS Total')
plt.show()
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(player_weight,PTS_count)
regress_values = player_weight * slope + intercept

plt.scatter(player_weight,PTS_count)
plt.plot(player_weight,regress_values, color='r')
plt.title("Weight vs. PTS")
plt.xlabel('Height (cm)')
plt.ylabel('PTS Total')
plt.show()
```
